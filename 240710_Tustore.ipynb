```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/tustore_ord.csv',encoding='cp949', sep=';')
df
```

```python
# 1. time (시간 컬럼 추가)
df['time'] = df['ord_time'].astype(str).str[:2].astype(int)
df
```

```python
# 2. 주간 야간 컬럼 추가
df['period'] = df['time'].apply(lambda x: '주간' if x < 18 else '야간')
df
```

```python
# 3. 날짜 생성 / 주말 유무
df['ord_dt'] = pd.to_datetime(df['ord_dt'].astype(str), format='%Y%m%d')

day_map = {'Monday': '월요일', 'Tuesday': '화요일', 'Wednesday': '수요일', 'Thursday': '목요일', 'Friday': '금요일', 'Saturday': '토요일', 'Sunday': '일요일'}
df['day_of_week'] = df['ord_dt'].dt.day_name().map(day_map)

def is_weekday(day, date):
    if date in ['2023-12-25', '2024-01-01']:
        return '주말'
    return '주중' if day in ['월요일', '화요일', '수요일', '목요일', '금요일'] else '주말'

df['is_weekday'] = df.apply(lambda x: is_weekday(x['day_of_week'], x['ord_dt'].strftime('%Y-%m-%d')), axis=1)
```

```python
# 4. 상품대분류 나누기
# 카테고리 정의
category_dict = {
    '스낵': ['캬라멜팝콘(M)', '츄러스', '캬라멜 팝콘(L)', '치즈핫도그', '후라이드치킨', '오리지널 팝콘(M)', '프레즐', '핫바', '하겐다즈 바닐라',
            '하겐다즈 쵸코', '오리지널 팝콘(L)', '하겐다즈 딸기', '스트로베리 조각케잌'],
    '음료': ['무알콜 맥주', '아이스 아메리카노', '밀키스', '생수', '코카콜라', '아이스티(복숭아)', '테라(캔)', '토레타', '스프라이트',
           '아메리카노(HOT)', '미닛메이드 포도', '핫초코', '허브티', '사케', '카페라떼', '미닛메이드 오렌지', '쵸코라떼 (HOT)', '하이볼'],
    '세트': ['Sweet ComboSet', 'Single Set', 'Beer set', 'Family set', 'Double ComboSet']
}

# 'category' 열 추가
def get_category(product):
    for category, products in category_dict.items():
        if product in products:
            return category
    return '기타'

df['B_category'] = df['goods_nm'].apply(get_category)
df
```

```python
category_dict = {
    '스낵': {
        '팝콘': ['캬라멜팝콘(M)', '캬라멜 팝콘(L)', '오리지널 팝콘(M)', '오리지널 팝콘(L)'],
        '디저트': ['하겐다즈 바닐라', '하겐다즈 쵸코', '하겐다즈 딸기', '스트로베리 조각케잌'],
        '스낵' : ['치즈핫도그','후라이드치킨','프레즐','핫바','츄러스']
    },
    '음료': {
        '주류': ['무알콜 맥주', '사케', '하이볼'],
        '음료': ['아이스 아메리카노', '밀키스', '생수', '코카콜라', '아이스티(복숭아)', '테라(캔)',
                  '토레타', '스프라이트', '아메리카노(HOT)', '미닛메이드 포도', '핫초코', '허브티', '카페라떼', '미닛메이드 오렌지', '쵸코라떼 (HOT)']
    },
    '세트': ['Sweet ComboSet', 'Single Set', 'Beer set', 'Family set', 'Double ComboSet']
}

# 'category' 열 추가
def get_category(product):
    for main_category, sub_categories in category_dict.items():
        if main_category == '스낵':
            for sub_category, products in sub_categories.items():
                if product in products:
                    return sub_category
        elif main_category == '음료':
            for sub_category, products in sub_categories.items():
                if product in products:
                    return sub_category
        elif main_category == '세트':
            if product in sub_categories:
                return '세트'
    return '기타'

df['M_category'] = df['goods_nm'].apply(get_category)
df
```
